```python
class counter:
    def __init__(self):
        self.value = 0
    def add_value(self):
        self.value += 1
    def subs_value(self):
        self.value -= 1
    def reset_counter(self):
        self.value = 0
    def get_value(self):
        print(self.value)
```

# zadanie 3

```python
import numpy as np
```

```python
class vector2D:
    def __init__(self, x, y):
        self.x=x
        self.y=y
    def __str__(self):
        return(f'({self.x}, {self.y})')
    def __add__(self, other):
        return vector2D(self.x + other.x, self.y + other.y)
    def __mul__(self, scalar):
        return vector2D(self.x*scalar, self.y*scalar)
    def __rmul__(self, scalar):
        return vector2D(self.x*scalar, self.y*scalar)
    def lenght(self):
        return np.sqrt(self.x^2+self.y^2)
    def scalar(self, other):
        return self.x*other.x + self.y*other.y
```

```python
#tworzymy 2 wektory
vector=vector2D(3, 7)
vector2=vector2D(1, 2)
#printujemy 2 wektory
print(vector)
print(vector2)
#liczymy sumę wektorów
print(vector + vector2)
#mnożymy wektor przez skalar
print(vector2 * 2)
print(8 * vector)
#długość wektora
print(vector.lenght())
#iloczyn skalarny wektorów
vector2D.scalar(vector, vector2)
```

# zadanie 4

```python
class measurement_series:
    def __init__(self, lista):
        self.lista=lista
    def __getitem__(self, i):
        return self.lista[i]
    def __setitem__(self, i, a):
        self.lista[i] = a
    def __len__(self):
        return len(self.lista)
    def __iter__(self):
        return iter(self.lista)
    def mean(self):
        return sum(self.lista)/len(self.lista)
    def przedzial(self):
        return min(self.lista), max(self.lista)
```

```python
lst=measurement_series([2, 5, 7, 9])
```

```python
lst[1] #pobieranie elementów przez []
lst[3]=0 #zmiana elementu przez []
len(lst) #liczba elementów
# iterowanie w pętli
for m in lst:
    print(m)
print(lst.mean()) # średnia
print(lst.przedzial()) # przedział (min, max)
```

# zadanie 5

```python
class Polynomial:
    def __init__(self, coeffs):
        self.coeffs=coeffs
    def __str__(self):
        terms=[]
        for p, coeff in enumerate(self.coeffs):
            if coeff == 0:
                continue
            if p == 0:
                terms.append(f'{coeff}')
            elif p == 1 and coeff < 0:
                terms.append(f'{coeff}x')
            elif p == 1 and coeff > 0:
                terms.append(f'+{coeff}x')
            elif p > 1 and coeff < 0:
                terms.append(f'{coeff}x^{p}')
            elif p > 1 and coeff > 0:
                terms.append(f'+{coeff}x^{p}')
        return ''.join(terms)
    def __call__(self, number):
        sum = 0
        for p, coeff in enumerate(self.coeffs):
            if p == 0:
                sum += coeff
            else:
                sum += (number)**p * coeff
        return sum
    def return_power(self):
        for i in range(len(self.coeffs)-1, -1, -1):
            if self.coeffs[i] != 0:
                return i
```

```python
h=Polynomial([1, 0, -2, 3, 0]) # tworzymy wielomian
print(h) # printujemy wielomian

print(h(2)) # obliczamy wartość wielomianu dla x=2
print(h.return_power()) #zwracamy stopień wielomianu
```

# zadanie 6

```python
class Account:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    def plus_balance(self, amount):
        self.balance += amount
    def minus_balance(self, amount):
        self.balance -= amount
    def balance_info(self):
        print(f'Użytkownik {self.name} ma na koncie {self.balance} punktów.')

class PremiumAccount(Account):
    def __init__(self, name, balance, bonus_factor):
        super().__init__(name, balance)
        self.bonus_factor = bonus_factor
    def plus_balance(self, amount):
        self.balance += amount * self.bonus_factor
        super().plus_balance(amount)
    def minus_balance(self, amount):
        super().minus_balance(amount)
    def balance_info(self):
        super().balance_info()
```

```python
my_account=Account('Ja', 140) #tworzymy konto zwykłe
my_account.plus_balance(30) #dodajemy punkty do konta
my_account.balance_info() #sprawdzamy stan konta
my_account.minus_balance(10) #odejmujemy punkty od konta
my_account.balance_info() #sprawdzamy stan konta
teacher_account=PremiumAccount('Maciej', 500, 0.2) #tworzymy konto premium
teacher_account.plus_balance(50) #dodajemy punkty do konta premium (powinno dodać 50 + 50*0.2 = 60)
teacher_account.balance_info() #sprawdzamy stan konta premium
```

```python
class Animal:
    def __init__(self, name, position):
        self.name=name
        self.position=position
    def move(self):
        pass
    def sound(self):
        pass
    def info(self):
        print(self.name, self.position)

class mrówka(Animal):
    def __init__(self, name, position):
        super().__init__(name, position)
    def sound(self):
        print('bzzzz')
        super().sound()
    def move(self):
        self.position[0]+=1
        self.position[1]+=1
        print(f'{self.name} przesunął się o {self.position[0]} w poziomie i {self.position[1]} w pionie')
        super().move()
        
class student(Animal):
    def __init__(self, name, position):
        super().__init__(name, position)
    def sound(self):
        print('Piwooo!')
        super().sound()
    def move(self):
        self.position[0]+=2
        self.position[1]+=2
        super().move()
        print(f'{self.name} przesunął się o {self.position[0]} w poziomie i {self.position[1]} w pionie')
```

```python
#tworzymy obiekty mrówki i studenta
Maciej=mrówka('Maciej', [0, 0])
Maciej.info()
Maciej.sound()
Maciej.move()

student=student('student', [0, 0])
student.info()
student.sound()
student.move()
```
