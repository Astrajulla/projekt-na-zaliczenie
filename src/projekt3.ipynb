```python
import time
```

```python
class TimeDecorator():
    def __init__(self, function = None, print_each_time=False):
        self.function = function
        self.times = []
        self.print_each_time = print_each_time
    def __call__(self, *args, **kwargs):
        if self.function is None:
            self.function = args[0]
            return self
        start_time=time.perf_counter()
        result=self.function(*args, **kwargs)
        end_time=time.perf_counter()
        measured_time=end_time-start_time
        self.times.append(measured_time)
        if self.print_each_time:
            print(f'(czas wykonania funkcji: {measured_time:.6f} s)')
        return result
    def statistics(self):
        print("STATYSTYKI:")
        return {
            "funkcja": self.function.__name__,
            "liczba pomiarów": len(self.times),
            "najdłuższy czas": max(self.times),
            "najkrótszy czas": min(self.times),
            "średni czas": sum(self.times) / len(self.times)
        }
```

```python
# testowanie funkcji z możliwoscią włączenia i wyłączenia wypisania czasu wykonania
@TimeDecorator(print_each_time=True)
def szukaj_liczby():
    time.sleep(0.5)
    return "znaleziono"

# Testowanie
szukaj_liczby()
szukaj_liczby()
szukaj_liczby()
print(szukaj_liczby.statistics())
```

```python
# testowanie funkcji obliczeniowej
@TimeDecorator()
def obliczenia(n):
    suma = 0
    for i in range(n):
        suma += i ** 3
    return suma

# Testowanie
obliczenia(1_000_000)
obliczenia(1_000_000)
obliczenia(1_000_000)
print(obliczenia.statistics())
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def celsius(self):
        return self._celsius
    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            self._celsius = -273.15
        else:
            self._celsius = value
    
    @property
    def fahrenheit(self):
        return self.celsius * 9 / 5 + 32
    @fahrenheit.setter
    def fahrenheit(self, value):
        if value < -459.67:
            self.celsius = -273.15
        else:
            self.celsius = (value - 32) * 5 / 9

    @property
    def kelvin(self):
        return self.celsius + 273.15
    @kelvin.setter
    def kelvin(self, value):
        if value < 0:
            self.celsius = -273.15
        else:
            self.celsius = value - 273.15

    def __str__(self):
        return f"{self.celsius:.2f} °C"

    @property
    def is_boiling(self):
        return self.celsius >= 100
    
    @property
    def is_freezing(self):
        return self.celsius <= 0
    
    @property
    def state(self):
        if self.celsius <= 0:
            return 'solid'
        elif self.celsius > 0 and self.celsius < 100:
            return 'fluid'
        elif self.celsius >= 100:
            return 'gas'    
```

```python
t = Temperature(25)
print(t.state)
```

```python
t = Temperature(25)

print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0
print(t.kelvin)       # 298.15

t.fahrenheit = 32
print(t.celsius)      # 0.0

t.kelvin = 300
print(t.celsius)      # 26.85

t.kelvin = -300      
print(t.kelvin)
```

```python
def generator(a, c, m, seed, N):
    x = seed
    for _ in range(N):
        x = (a * x + c) % m
        yield x
```

```python
# generowanie przy użyciu pętli for
gen = generator(seed=1, a=5, c=3, m=16, N=4)
for l in gen:
    print(l)
```

```python
# generowanie przy użyciu funkcji next()
gen = generator(seed=1, a=5, c=3, m=16, N=4)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#generator liczb zmiennoprzecinkowych
def float_generator(a, c, m, seed, N):
    x = seed
    for _ in range(N):
        x = (a * x + c) % m
        yield x/m
```

```python
# generowanie przy użyciu funkcji next()
gen2 = float_generator(seed=1, a=5, c=3, m=16, N=4)
print(next(gen2))
print(next(gen2))
print(next(gen2))
print(next(gen2))
```

```python
# generowanie przy użyciu pętli for
gen = float_generator(seed=1, a=5, c=3, m=16, N=4)
for i in gen:
    print(i)
```
